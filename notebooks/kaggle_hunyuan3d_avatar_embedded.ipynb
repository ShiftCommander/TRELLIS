# Hunyuan3D Avatar 3D (Kaggle embedded image)

Kaggle T4 pipeline: embedded input image, Hunyuan3D-2mini shape, then Hunyuan3D-Paint texture.


```python
import shutil
import subprocess

if shutil.which('nvidia-smi') is None:
    raise RuntimeError('GPU not detected. In Kaggle Settings, set Accelerator to GPU and restart the session.')
subprocess.run(['nvidia-smi'], check=False)
```

```python
import base64
from pathlib import Path

WORKDIR = Path('/kaggle/working')
WORKDIR.mkdir(parents=True, exist_ok=True)
INPUT_IMAGE_PATH = WORKDIR / 'avatar-man-1.png'
AVATAR_MAN_1_PNG_B64 = ''
INPUT_IMAGE_PATH.write_bytes(base64.b64decode(AVATAR_MAN_1_PNG_B64))
SELECTED_IMAGE = str(INPUT_IMAGE_PATH)
print(f'Image written: {SELECTED_IMAGE} ({INPUT_IMAGE_PATH.stat().st_size} bytes)')
```

```python
INPUT_IMAGE = SELECTED_IMAGE
OUT_DIR = '/kaggle/working/outputs/avatar-man-1-hunyuan'
SEED = 12345
OCTREE_RESOLUTION = 320

# T4 presets:
# MAX_QUALITY: 35 / 20000
# BALANCED:    28 / 12000
# SAFE:        24 / 8000
NUM_INFERENCE_STEPS = 28
NUM_CHUNKS = 12000
```

```python
import shutil
import subprocess
from pathlib import Path

REPO_DIR = Path('/kaggle/working/Hunyuan3D-2')
ZIP_PATH = Path('/kaggle/working/Hunyuan3D-2-main.zip')
ZIP_URL = 'https://codeload.github.com/Tencent-Hunyuan/Hunyuan3D-2/zip/refs/heads/main'

def run(cmd: str, check: bool = True) -> bool:
    print(f'\n$ {cmd}')
    result = subprocess.run(cmd, shell=True)
    if check and result.returncode != 0:
        raise RuntimeError(f'Command failed: {cmd}')
    return result.returncode == 0

if REPO_DIR.exists():
    shutil.rmtree(REPO_DIR)

if not ZIP_PATH.exists():
    run('python3 -m pip install -q requests')
    import requests
    r = requests.get(ZIP_URL, timeout=120)
    r.raise_for_status()
    ZIP_PATH.write_bytes(r.content)
    print(f'Downloaded {ZIP_PATH} ({ZIP_PATH.stat().st_size} bytes)')

run(f'unzip -q "{ZIP_PATH}" -d /kaggle/working')
extracted = Path('/kaggle/working/Hunyuan3D-2-main')
if not extracted.exists():
    raise RuntimeError('Zip extracted but /kaggle/working/Hunyuan3D-2-main was not found.')
extracted.rename(REPO_DIR)

run('cd /kaggle/working/Hunyuan3D-2 && python3 -m pip install --upgrade pip')
run('cd /kaggle/working/Hunyuan3D-2 && python3 -m pip install torch==2.4.0 torchvision==0.19.0 --index-url https://download.pytorch.org/whl/cu121')
run('cd /kaggle/working/Hunyuan3D-2 && python3 -m pip install -r requirements.txt')
run('cd /kaggle/working/Hunyuan3D-2 && python3 -m pip install diffusers==0.29.2')
run('cd /kaggle/working/Hunyuan3D-2 && python3 -m pip install -e .')

run('cd /kaggle/working/Hunyuan3D-2/hy3dgen/texgen/custom_rasterizer && python3 setup.py install')
run('cd /kaggle/working/Hunyuan3D-2/hy3dgen/texgen/differentiable_renderer && python3 setup.py install')

import torch, diffusers
print('torch', torch.__version__)
print('diffusers', diffusers.__version__)
print('Environment ready.')
```

```python
import gc
import json
import os
import shutil
import time
from pathlib import Path

os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

import torch
from PIL import Image

os.chdir('/kaggle/working/Hunyuan3D-2')

from hy3dgen.rembg import BackgroundRemover
from hy3dgen.shapegen import Hunyuan3DDiTFlowMatchingPipeline

out_dir = Path(OUT_DIR)
out_dir.mkdir(parents=True, exist_ok=True)

raw = Image.open(INPUT_IMAGE)
image = raw.convert('RGBA')
if raw.mode == 'RGB':
    image = BackgroundRemover()(image)

started = time.time()
shape_pipeline = Hunyuan3DDiTFlowMatchingPipeline.from_pretrained(
    'tencent/Hunyuan3D-2mini',
    subfolder='hunyuan3d-dit-v2-mini',
    variant='fp16',
)
shape_pipeline.enable_model_cpu_offload()

mesh = shape_pipeline(
    image=image,
    num_inference_steps=NUM_INFERENCE_STEPS,
    octree_resolution=OCTREE_RESOLUTION,
    num_chunks=NUM_CHUNKS,
    generator=torch.manual_seed(SEED),
    output_type='trimesh',
)[0]

shape_glb = out_dir / 'avatar-man-1_hunyuan_shape.glb'
mesh.export(shape_glb)

del shape_pipeline
torch.cuda.empty_cache()
gc.collect()

texture_status = 'SKIPPED'
final_glb = shape_glb
texture_error = None
try:
    from hy3dgen.texgen import Hunyuan3DPaintPipeline
    paint_pipeline = Hunyuan3DPaintPipeline.from_pretrained('tencent/Hunyuan3D-2')
    paint_pipeline.enable_model_cpu_offload()
    textured_mesh = paint_pipeline(mesh, image=image)
    final_glb = out_dir / 'avatar-man-1_hunyuan_textured.glb'
    textured_mesh.export(final_glb)
    texture_status = 'PASS'
except Exception as exc:
    texture_status = 'FAIL'
    texture_error = f'{type(exc).__name__}: {exc}'

manifest = {
    'pipeline': 'Hunyuan3D-2mini + Hunyuan3D-Paint',
    'input_image': INPUT_IMAGE,
    'seed': SEED,
    'octree_resolution': OCTREE_RESOLUTION,
    'num_inference_steps': NUM_INFERENCE_STEPS,
    'num_chunks': NUM_CHUNKS,
    'shape_glb': str(shape_glb),
    'final_glb': str(final_glb),
    'texture_status': texture_status,
    'texture_error': texture_error,
    'elapsed_seconds': round(time.time() - started, 2),
}
(out_dir / 'manifest.json').write_text(json.dumps(manifest, indent=2), encoding='utf-8')

delivery_dir = out_dir / 'avatar-man-1_hunyuan_delivery'
if delivery_dir.exists():
    shutil.rmtree(delivery_dir)
delivery_dir.mkdir(parents=True, exist_ok=True)
shutil.copy2(out_dir / 'manifest.json', delivery_dir / 'manifest.json')
shutil.copy2(final_glb, delivery_dir / final_glb.name)
if final_glb != shape_glb:
    shutil.copy2(shape_glb, delivery_dir / shape_glb.name)

zip_path = shutil.make_archive(str(out_dir / 'avatar-man-1_hunyuan_delivery'), 'zip', root_dir=str(delivery_dir))
print(json.dumps({'manifest': manifest, 'zip': zip_path}, indent=2))
if texture_status != 'PASS':
    raise RuntimeError(f'Texture generation did not pass: {texture_error}')
```
